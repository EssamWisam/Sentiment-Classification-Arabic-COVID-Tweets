```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, HTML
from collections import Counter

plt.rcParams['figure.dpi'] = 300
plt.rcParams.update({'font.size': 4})


# Training Set
training_set = pd.read_csv('../../Dataset/train.csv')
training_set.category = pd.Categorical(training_set.category)
training_set['category'] = training_set.category.cat.codes
# ['advice: 0' 'celebrity:1' 'info_news:2' 'others:3' 'personal:4' 'plan:5' 'requests:6', 'restrictions:7' 'rumors:8' 'unrelated:9']
display(training_set)


x ,y1, y2 = training_set['text'].values,training_set['stance'].values, training_set['category']

# Validation Set
dev_set = pd.read_csv('../../Dataset/dev.csv')
dev_set.category = pd.Categorical(dev_set.category)
dev_set['category'] = dev_set.category.cat.codes
display(dev_set)

X ,Y1, Y2 = dev_set['text'].values,dev_set['stance'].values, dev_set['category']

# make histogram of the stance

data = [y1, y2, Y1, Y2]
titles = ['Training Stance','Training Categories','Testing Stance','Testing Categories'] 

f,a = plt.subplots(2,2)
a = a.ravel()
for idx,ax in enumerate(a):
    counts, edges, bars = ax.hist(data[idx])
    ax.bar_label(bars)
    ax.set_title(titles[idx])

plt.tight_layout()
```
